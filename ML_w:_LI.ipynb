```python
#Elastic net with transformations for lasso-selected features
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

numeric_cols = []
for col in cv_data_gpa_important1.columns:
    if cv_data_gpa_important1[col].dtype == 'float64' and cv_data_gpa_important1[col].nunique() > 30:
        numeric_cols.append(col)
categorical_cols = [col for col in cv_data_gpa_important1.columns if col not in numeric_cols]

def safe_log_transform(x):
    x_min_positive = x - np.min(x) + 0.01  
    return np.log(x_min_positive)

def safe_sqrt_transform(x):
    x_abs = np.abs(x)
    return np.sqrt(x_abs)

def squared_transform(x):
    return np.square(x)

# Initialize FunctionTransformers using the safe functions
log_transform = FunctionTransformer(safe_log_transform, validate=False)
sqrt_transform = FunctionTransformer(safe_sqrt_transform, validate=False)
squared_transform = FunctionTransformer(squared_transform, validate=False)

# Initialize a dictionary to store scalers for each column
scalers = {}

# Apply transformations and scale features for the training set
transformed_features1 = pd.DataFrame(index=cv_data_gpa_important1.index)
for col in numeric_cols:
    transformed_data = np.hstack([
        log_transform.transform(cv_data_gpa_important1[[col]] + 1),
        sqrt_transform.transform(cv_data_gpa_important1[[col]] + 1),
        squared_transform.transform(cv_data_gpa_important1[[col]])
    ])
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(transformed_data)
    scalers[col] = scaler
    for i, transform_name in enumerate(['log', 'sqrt', 'squared']):
        transformed_features1[f'{col}_{transform_name}'] = scaled_data[:, i]

transformed_features1 = pd.concat([transformed_features1, cv_data_gpa_important1[categorical_cols]], axis=1)

# Fit Elastic Net model (using GridSearchCV)
param_grid = {
    'l1_ratio': np.arange(0.0, 1.05, 0.05)}

elastic_net = ElasticNet(random_state=123)
grid_search = GridSearchCV(elastic_net, param_grid=param_grid, cv=10, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(transformed_features1, labels)
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Best MSE from CV results
best_mse_cv = -grid_search.best_score_
print(f"Best Mean Squared Error on CV set: {best_mse_cv:.4f}")

# Prepare the test set transformations
transformed_test_features1 = pd.DataFrame(index=test_data_gpa_important1.index)

for col in numeric_cols:
    transformed_data_test = np.hstack([
        log_transform.transform(test_data_gpa_important1[[col]] + 1), 
        sqrt_transform.transform(test_data_gpa_important1[[col]] + 1), 
        squared_transform.transform(test_data_gpa_important1[[col]])
    ])
    scaler = scalers[col]
    scaled_data_test = scaler.transform(transformed_data_test)
    for i, transform_name in enumerate(['log', 'sqrt', 'squared']):
        transformed_test_features1[f'{col}_{transform_name}'] = scaled_data_test[:, i]
transformed_test_features1 = pd.concat([transformed_test_features1, test_data_gpa_important1[categorical_cols]], axis=1)

# Fit and predict using Elastic Net with the best parameters
elastic_net_best = ElasticNet(l1_ratio=best_params['l1_ratio'], random_state=123)
elastic_net_best.fit(transformed_features1,labels)
test_predictions = elastic_net_best.predict(transformed_test_features1)

# Calculate MSE on the test set
test_mse = mean_squared_error(test_labels_gpa, test_predictions)
print(f"Mean Squared Error on Test Set: {test_mse:.4f}")
```
